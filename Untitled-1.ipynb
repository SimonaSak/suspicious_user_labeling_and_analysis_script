```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# 1. DUOMENŲ UŽKROVIMAS
# ============================================================

features = pd.read_excel('Outputs/final_features.xlsx')
iforest_suspicious
lof_suspicious

# Užtikriname, kad account_number tipas visur sutaptų
features['account_number'] = features['account_number'].astype(str)
iforest_suspicious['account_number'] = iforest_suspicious['account_number'].astype(str)
lof_suspicious['account_number'] = lof_suspicious['account_number'].astype(str)

# ============================================================
# 2. ANALIZUOJAMI POŽYMIAI
# ============================================================

FEATURE_COLS = [
    'total_calls',
    'mean_call_duration',
    'unique_contacts_called',
    'max_calls_to_single_contact',
    'failed_calls_count',
    'failed_call_share',
    'answered_call_share',
    'night_call_share',
    'active_day_ratio',
    'avg_calls_per_active_day',
    'night_calls_count',
    'unique_destination_codes',
    'unique_numbers_per_account',
    'webrtc_call_share'
]

FEATURE_LABELS = {
    'total_calls': 'Bendras skambučių sk.',
    'mean_call_duration': 'Vid. skambučio trukmė',
    'unique_contacts_called': 'Unikalūs kontaktai',
    'max_calls_to_single_contact': 'Maks. sk. vienam kontaktui',
    'failed_calls_count': 'Nesėkmingi skambučiai',
    'failed_call_share': 'Nesėkmingų sk. dalis',
    'answered_call_share': 'Atsilieptų sk. dalis',
    'night_call_share': 'Naktinių sk. dalis',
    'active_day_ratio': 'Aktyvių dienų santykis',
    'avg_calls_per_active_day': 'Vid. sk./aktyvią dieną',
    'night_calls_count': 'Naktinių skambučių sk.',
    'unique_destination_codes': 'Unikalūs krypties kodai',
    'unique_numbers_per_account': 'Unikalūs numeriai paskyroje',
    'webrtc_call_share': 'WebRTC sk. dalis'
}

# Pasiliekame tik tuos požymius, kurie tikrai yra faile
FEATURE_COLS = [col for col in FEATURE_COLS if col in features.columns]

# ============================================================
# 3. POPULIACIJA IR Q85+ GRUPĖS
# ============================================================

population = features[['account_number'] + FEATURE_COLS].copy()

iforest_q85_ids = iforest_suspicious.loc[
    iforest_suspicious['risk_group'].str.contains('Aukšta', case=False, na=False),
    'account_number'
]

lof_q85_ids = lof_suspicious.loc[
    lof_suspicious['risk_group'].str.contains('Aukšta', case=False, na=False),
    'account_number'
]

iforest_q85 = features[features['account_number'].isin(iforest_q85_ids)][FEATURE_COLS]
lof_q85 = features[features['account_number'].isin(lof_q85_ids)][FEATURE_COLS]

print("Populiacijos vartotojų skaičius:", len(population))
print("iForest Q85+ vartotojų skaičius:", len(iforest_q85))
print("LOF Q85+ vartotojų skaičius:", len(lof_q85))

# ============================================================
# 4. FUNKCIJA PROFILIO PALYGINIMUI
# ============================================================

def compare_group_to_population(group_df, population_df, group_name):
    rows = []

    for feature in FEATURE_COLS:
        pop_median = population_df[feature].median()
        group_median = group_df[feature].median()

        pop_mean = population_df[feature].mean()
        group_mean = group_df[feature].mean()

        median_diff = group_median - pop_median
        mean_diff = group_mean - pop_mean

        median_ratio = np.nan if pop_median == 0 else group_median / pop_median
        mean_ratio = np.nan if pop_mean == 0 else group_mean / pop_mean

        rows.append({
            'Požymis': FEATURE_LABELS.get(feature, feature),
            'Populiacijos mediana': pop_median,
            f'{group_name} mediana': group_median,
            'Medianų skirtumas': median_diff,
            'Medianų santykis': median_ratio,
            'Populiacijos vidurkis': pop_mean,
            f'{group_name} vidurkis': group_mean,
            'Vidurkių skirtumas': mean_diff,
            'Vidurkių santykis': mean_ratio
        })

    result = pd.DataFrame(rows)

    # Patogesnis rikiavimas pagal tai, kur grupė labiausiai skiriasi nuo populiacijos
    result['Nukrypimas nuo populiacijos'] = abs(result['Medianų santykis'] - 1)
    result = result.sort_values('Nukrypimas nuo populiacijos', ascending=False)

    return result

iforest_profile = compare_group_to_population(
    iforest_q85,
    population,
    'iForest Q85+'
)

lof_profile = compare_group_to_population(
    lof_q85,
    population,
    'LOF Q85+'
)

# ============================================================
# 5. REZULTATŲ IŠVEDIMAS
# ============================================================

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
pd.set_option('display.float_format', '{:.3f}'.format)

print("\n" + "="*100)
print("iForest Q85+ aukštos rizikos vartotojų profilis, palyginti su populiacija")
print("="*100)
display(iforest_profile)

print("\n" + "="*100)
print("LOF Q85+ aukštos rizikos vartotojų profilis, palyginti su populiacija")
print("="*100)
display(lof_profile)

# ============================================================
# 6. TOP POŽYMIAI, KURIE LABIAUSIAI IŠSKIRIA Q85+ GRUPES
# ============================================================

print("\nLabiausiai nuo populiacijos besiskiriantys iForest Q85+ požymiai:")
display(iforest_profile[['Požymis', 'Populiacijos mediana', 'iForest Q85+ mediana', 'Medianų santykis']].head(7))

print("\nLabiausiai nuo populiacijos besiskiriantys LOF Q85+ požymiai:")
display(lof_profile[['Požymis', 'Populiacijos mediana', 'LOF Q85+ mediana', 'Medianų santykis']].head(7))

# ============================================================
# 7. VIZUALIZACIJA: MEDIANŲ SANTYKIS SU POPULIACIJA
# ============================================================

plot_df = pd.DataFrame({
    'Požymis': [FEATURE_LABELS.get(f, f) for f in FEATURE_COLS],
    'iForest Q85+ / populiacija': [
        iforest_q85[f].median() / population[f].median() if population[f].median() != 0 else np.nan
        for f in FEATURE_COLS
    ],
    'LOF Q85+ / populiacija': [
        lof_q85[f].median() / population[f].median() if population[f].median() != 0 else np.nan
        for f in FEATURE_COLS
    ]
})

plot_df = plot_df.replace([np.inf, -np.inf], np.nan).dropna()

x = np.arange(len(plot_df))
width = 0.38

plt.figure(figsize=(15, 7))
plt.bar(x - width/2, plot_df['iForest Q85+ / populiacija'], width, label='iForest Q85+')
plt.bar(x + width/2, plot_df['LOF Q85+ / populiacija'], width, label='LOF Q85+')

plt.axhline(1, linestyle='--', linewidth=1, color='black')
plt.xticks(x, plot_df['Požymis'], rotation=45, ha='right')
plt.ylabel('Medianų santykis su populiacija')
plt.title('Aukštos rizikos vartotojų Q85+ profilio palyginimas su populiacija')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

```
